```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Caption', 'Footnote', 'Formula', 'List-item', 'Page-footer', 'Page-header', 'Picture', 'Section-hea..',
              'Table', 'Text', 'Title']
numbers = [20981, 5931, 23061, 175138, 66884, 54656, 42442, 134334, 32339, 480443, 4736]

fig, ax = plt.subplots()
y_pos = np.arange(len(categories))
ax.barh(y_pos, numbers, align='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(categories)

ax.set_xlabel('Number of Occurrences')

plt.title('Bar Plot of Categories')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Caption', 'Footnote', 'Formula', 'List-item', 'Page-footer', 'Page-header', 'Picture', 'Section-hea..',
              'Table', 'Text', 'Title']
numbers = [20981, 5931, 23061, 175138, 66884, 54656, 42442, 134334, 32339, 480443, 4736]

sorted_data = sorted(zip(categories, numbers), key=lambda x: x[1], reverse=True)
categories, numbers = zip(*sorted_data)

fig, ax = plt.subplots()
y_pos = np.arange(len(categories))
ax.barh(y_pos, numbers, align='center')

ax.set_yticks(y_pos)
ax.set_yticklabels(categories)

ax.set_xlabel('Number of Occurrences')

plt.title('Bar Plot of Classes Within The Documents')

plt.show()
```
